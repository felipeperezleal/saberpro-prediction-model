###Codificación de Variables Categóricas

Necesitamos convertir variables categóricas a formato numérico para que puedan ser utilizadas por nuestro modelo. Utilizaremos las siguientes técnicas de codificación:

1.   **One-Hot Encoding:** Para variables no ordinales como 'ESTU_PRGM_ACADEMICO', 'ESTU_PRGM_DEPARTAMENTO', etc.
2.   **Label Encoding:** Para variables ordinales como 'RENDIMIENTO_GLOBAL'.

Imprimamos los valores únicos para cada columna en el conjunto de datos junto con la cantidad de veces que aparece cada valor. Así visualizaremos y analizaremos los valores presentes en cada columna para tomar una decisión sobre cómo manejar y codificar cada variable.

````python
for column in train.columns:
    unique_values = train[column].unique()
    print(f"Columna: {column}")
    for value in unique_values:
        count = (train[column] == value).sum()
        print(f"  Valor: {value} - Apariciones: {count}")
    print("\n")
````

Para las variables binarias, aplicaremos la codificación de etiquetas utilizando el LabelEncoder de Scikit-Learn para convertirlas en valores numéricos binarios (0 o 1).

````python
from sklearn.preprocessing import LabelEncoder

binary_columns = ['FAMI_TIENEINTERNET',
                  'FAMI_TIENELAVADORA',
                  'FAMI_TIENEAUTOMOVIL',
                  'ESTU_PRIVADO_LIBERTAD',
                  'ESTU_PAGOMATRICULAPROPIO',
                  'FAMI_TIENECOMPUTADOR'
]

label_encoder = LabelEncoder()

for column in binary_columns:
    train[column] = label_encoder.fit_transform(train[column])
    test[column] = label_encoder.fit_transform(test[column])
````

Para las variables como "ESTU_PRGM_ACADEMICO"

Se creará una categoría "Otros" debido a que contiene datos con múltiples categorías, algunas de las cuales pueden tener pocas instancias. Así simplificamos y generalizamos los datos, agrupando las categorías menos comunes bajo una etiqueta común.

````python
#Combinamos train y test para realizar la agrupación de programas
combined = pd.concat([train, test])

#Encuentra los programas menos comunes en el conjunto combinado
program_counts = combined['ESTU_PRGM_ACADEMICO'].value_counts()
threshold = 500
less_common = program_counts[program_counts < threshold].index.tolist()

#Aplica la agrupación en el conjunto combinado
combined['ESTU_PRGM_ACADEMICO'] = combined['ESTU_PRGM_ACADEMICO'].apply(
    lambda x: 'Otros' if x in less_common else x
)

#Dividimos nuevamente en train y test
train = combined.iloc[:len(train)]
test = combined.iloc[len(train):]

#Como test no debería tener la variable a predecir la dropeamos, igualmente es una columna de NaN
test.drop('RENDIMIENTO_GLOBAL', axis=1, inplace=True)
````

Al concatenar los conjuntos de datos (train y test) en combined, se combinan verticalmente (uno encima del otro) pero las filas y sus valores se mantienen independientes. La combinación se hace para realizar operaciones comunes a ambos conjuntos, finalmente, se separan nuevamente los conjuntos originales (train y test) para continuar trabajando con ellos por separado.

No queremos que nuestro modelo asuma una relación de orden entre los períodos, evitando cualquier sesgo que podría surgir de interpretar los valores numéricos como una secuencia.

Entonces, para la columna "PERIODO", aunque los valores puedan parecer numéricos, no hay una relación de orden natural entre ellos, ya que representan períodos de tiempo y no tienen una secuencia ordinal.

Al aplicar One-Hot Encoding a la columna "PERIODO", cada valor único se convertirá en una nueva columna binaria. Cada fila tendrá un 1 en la columna correspondiente al período al que pertenece y 0 en las demás columnas, lo que permite representar la presencia o ausencia de un período específico para cada entrada de datos.

````python
train = pd.get_dummies(train, columns=['PERIODO'])
test = pd.get_dummies(test, columns=['PERIODO'])
````

En columnas como 'FAMI_EDUCACIONMADRE' o 'FAMI_EDUCACIONPADRE', las diferentes categorías representan distintos niveles educativos, los cuales pueden ordenarse de menor a mayor en términos de logros educativos. Similarmente, 'FAMI_ESTRATOVIVIENDA' o 'ESTU_VALORMATRICULAUNIVERSIDAD' tienen niveles que siguen una jerarquía.

````python
ordinal_encode = [
    'FAMI_EDUCACIONMADRE',
    'FAMI_EDUCACIONPADRE',
    'FAMI_ESTRATOVIVIENDA',
    'ESTU_VALORMATRICULAUNIVERSIDAD',
    'ESTU_HORASSEMANATRABAJA'
]

label_encoder = LabelEncoder()

for column in ordinal_encode:
    if column in train.columns:
        train[column] = label_encoder.fit_transform(train[column].astype(str))
        test[column] = label_encoder.fit_transform(test[column].astype(str))
````

Para las columnas 'ESTU_PRGM_ACADEMICO' y 'ESTU_PRGM_DEPARTAMENTO' que no presentan un orden intrínseco entre las categorías, utilizaremos el One-Hot Encoding.

Creando nuevas columnas binarias (0 o 1) para cada categoría única en estas columnas, permitiendo representar la presencia o ausencia de cada categoría en cada fila de datos.

````python
non_ordinal_encode = ['ESTU_PRGM_ACADEMICO', 'ESTU_PRGM_DEPARTAMENTO']

train = pd.get_dummies(train, columns=non_ordinal_encode)
test = pd.get_dummies(test, columns=non_ordinal_encode)
````

Para la variable 'RENDIMIENTO_GLOBAL', dado que representa diferentes niveles o categorías ordinales,deberíamos aplicar una codificación ordinal, asignando valores numéricos a cada categoría en función de su orden

````python
train['RENDIMIENTO_GLOBAL'].unique()
````

````python
label_encoder_rendimiento_global = LabelEncoder()
label_encoder_rendimiento_global.fit(train['RENDIMIENTO_GLOBAL'])

train['RENDIMIENTO_GLOBAL'] = label_encoder.fit_transform(train['RENDIMIENTO_GLOBAL'])
````

Según el mapa de calor, observamos que la mayoría de características influyentes en el rendimiento de la prueba de los estudiantes, se encuentra en la parte superior izquierda, es decir, en las características que no corresponden al departamento, periodo y (aunque está omitido en el mapa de calor para fines visualizativos) el programa académico

````python
exclude_columns = ['ESTU_PRGM_ACADEMICO', 'ESTU_PRGM_DEPARTAMENTO', 'RENDIMIENTO_GLOBAL']
scatter_columns = [col for col in train.columns if not any(col.startswith(exclude_col) for exclude_col in exclude_columns)]

correlations = {}
for feature1 in scatter_columns:
    correlation = train[feature1].corr(train['RENDIMIENTO_GLOBAL'])
    correlations[feature1] = correlation

#Ordena las características según su correlación con 'RENDIMIENTO_GLOBAL'
sorted_features = sorted(correlations, key=lambda x: abs(correlations[x]), reverse=True)

num_cols = 3  #Número de columnas en la visualización
num_rows = (len(sorted_features) + num_cols - 1) // num_cols  #Cálculo del número de filas

fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 15))  #Crear subplots

for i, feature1 in enumerate(sorted_features):
    row = i // num_cols
    col = i % num_cols

    axs[row, col].scatter(train[feature1], train['RENDIMIENTO_GLOBAL'], alpha=0.5)
    axs[row, col].set_title(f'{feature1} vs RENDIMIENTO_GLOBAL')
    axs[row, col].set_xlabel(feature1)
    axs[row, col].set_ylabel('RENDIMIENTO_GLOBAL')
    axs[row, col].grid(True)

#Oculta los subplots no utilizados si hay más subplots que características a mostrar
for i in range(len(sorted_features), num_rows * num_cols):
    row = i // num_cols
    col = i % num_cols
    axs[row, col].axis('off')

plt.tight_layout()
plt.show()
````

El gráfico nos permite visualizar cómo se relaciona una característica respecto a la característica objetivo 'RENDIMIENTO_GLOBAL'

Visualizando los gráficos vemos una homogeneidad en todos los casos, entre los posibles valores que toma cada columna y la objetivo (el rendimiento global)

#E. Exploración de hiperparámetros

Probemos diferentes tasas de aprendizaje (learning rates) en nuestro modelo, encontrando la mejor tasa, ajustamos la magnitud de los ajustes que se hacen a los pesos de la red neuronal durante el proceso de entrenamiento

````python
lr = [0.1, 0.01, 0.001, 0.0001, 0.00001]
report = []
for r in lr:
  model = Sequential()
  model.add(Dense(264, input_dim=X.shape[1], activation='relu'))
  model.add(Dense(256, activation='relu'))
  model.add(Dense(y.shape[1], activation='softmax'))
  opt = Adam(learning_rate=r)
  model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
  h = model.fit(x=X_train, y=y_train, batch_size=8192, epochs=20, validation_data=(X_val, y_val), verbose=1)
  report.append(h)
````

Veamos la diferencia entre las curvas de aprendizaje dependiendo de cada tasa de aprendizaje

````python
plt.figure(figsize=(12,12))
plt.subplot(2,1,1)
for i in range(len(report)):
  plt.plot(report[i].history['loss'], label='Entrenamiento - lr = {}'.format(lr[i]))
plt.ylabel("Pérdida en entrenamiento")
plt.title("Curva de aprendizaje - pérdida en entrenamiento")
plt.legend()
plt.subplot(2,1,2)
for i in range(len(report)):
  plt.plot(report[i].history['val_loss'], label='Validación - lr = {}'.format(lr[i]))
plt.ylabel("Pérdida en validación")
plt.title("Curva de aprendizaje - pérdida en validación")
plt.legend()

plt.tight_layout()
plt.show()
````

La línea para lr=0.1 disminuye rápidamente al principio, pero luego se estabiliza a un nivel relativamente alto. Las líneas para lr=0.01, lr=0.001 y lr=0.0001 disminuyen de forma más gradual y alcanzan niveles de pérdida más bajos.

Estas diferencias indican que el modelo es más sensible a la tasa de aprendizaje. Con una tasa de aprendizaje demasiado alta, el modelo puede sobreajustar los datos de entrenamiento y, por lo tanto, no generalizar bien a los datos nuevos. Con una tasa de aprendizaje demasiado baja, el modelo puede tardar demasiado en aprender y, por lo tanto, puede no alcanzar su máximo rendimiento.

La línea para Ir=0.0001 muestra que el modelo podría alcanzar una pérdida de validación máss baja que la de las tasas lr=0.1, lr=0.01 e lr=0.001.

Precision y recall en el problema multiclase

Recordemos que:

* $Precision = \frac{TP}{TP + FP}$
* $Recall = \frac{TP}{TP + FN}$
* $F_1 = 2 * \frac{PRE*REC}{PRE + REC}$

##Validación Cruzada

````python
def getModelCrossVal():
    model_cross_val = Sequential()
    model_cross_val.add(Dense(264, input_dim=X.shape[1], activation='relu'))
    model_cross_val.add(Dense(256, activation='relu'))
    model_cross_val.add(Dense(y.shape[1], activation='softmax'))

    opt = Adam(learning_rate=0.00001)
    model_cross_val.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
    return model_cross_val

Kmodel = KerasClassifier(build_fn=getModelCrossVal, verbose=1, epochs=5, batch_size=32)
````

````python
#Definición de validación cruzada k-fold
kf = KFold(n_splits=2, shuffle=True)

scoring = ['precision_macro', 'recall_macro', 'roc_auc']

#Realizar validación cruzada y obtener las métricas
cv_results = cross_validate(Kmodel, X, y, scoring=scoring, cv=kf)


print("Precision (macro): {:.2f}".format(cv_results['test_precision_macro'].mean()))
print("Recall (macro): {:.2f}".format(cv_results['test_recall_macro'].mean()))
print("AUC-ROC: {:.2f}".format(cv_results['test_roc_auc'].mean()))
````

Basados en los resultados de la validación cruzada k-fold, podemos inferir lo siguiente:

Rendimiento general:

- La precisión macro promedio es de 0.38, lo que indica que el modelo clasifica correctamente el 38% de las instancias en promedio.

- El recall macro promedio es de 0.41, lo que indica que el modelo identifica correctamente el 41% de las instancias pertenecientes a cada clase en promedio

- El AUC-ROC promedio es de 0.68, lo que indica un rendimiento bueno/moderado del modelo para diferenciar entre clases.
